```python
opc2=True
try:

    while(opc2!=False):
        print("Selector de Faces")
        print("0 :Salir")
        
        for i, ing in enumerate(date3):
            print(f"{i+1} :{ing["name"]}")

        opc=int(input("Dame tu opcion: "))
        for ii, images in enumerate(date3):
            if(ii+1==opc):
                print(f"Has seleccionado a: {images["name"]}")
                x=images["image"]

                #Esto es todo lo que se necesita para crear la imagen 
                res=rq.get(x)
                res.raise_for_status()# Esto es lo que se le agrego
                imagen=Image.open(BytesIO(res.content))
                imagen.show()


        print("---------------------------------------")

        if(opc==0):
            opc2 = False
except ValueError:
    print("A ocurrido unerror")
    
```
